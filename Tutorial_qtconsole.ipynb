# How to open separate console in Jupyter Notebook
Sometimes we wish to have a separate console to look up variables, try a small code segment, or check if kernel is working properly. Having a separate console in addition to the notebook can be very handy.

## The code to do this is "%qtconsole"

```python
%qtconsole
```

It should look like this:


## A couple of simple demos

```python
a = 10
```

## A few things to pay attention to

1. You may open multiple consoles if you wish
2. If you close the console, your kernel will die
3. You can zoom in/out for better view
4. No auto completion when typing code
5. Arrays, dataframes, etc will display in a messy way. So it's not a good way to visualize large arrays and dataframes
